```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    confusion_matrix)
from sklearn.feature_selection import (
    SelectKBest,
    chi2
)
```

```python
df = pd.read_csv('/content/drive/MyDrive/CursoIA/4.Tópicos Avançados de Machine Learning/ad.data', header=None)
df.shape
```

```python
target = df.columns.tolist()[-1]
features = df.columns.tolist()[:-1]
y = df[target]
X = df[features]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=43)
```

```python
nb1 = GaussianNB()
nb1.fit(X_train, y_train)
```

```python
y_predict = nb1.predict(X_test)
acc_nb1 = accuracy_score(y_test, y_predict)
acc_nb1
```

```python
selecao = SelectKBest(chi2, k=7)
X_novo = selecao.fit_transform(X, y)
pd.DataFrame(X_novo).head()
```

```python
X_novo.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_novo, y, test_size=0.2, random_state=43)
```

```python
nb2 = GaussianNB()
nb2.fit(X_train, y_train)

y_predict = nb2.predict(X_test)
acc_nb2 = accuracy_score(y_test, y_predict)
acc_nb2
```

```python
y_test.value_counts()
```

```python
acc_nb2 = accuracy_score(y_test, y_predict)
precision_nb2 = precision_score(y_test, y_predict, pos_label='ad.')
recall_nb2 = recall_score(y_test, y_predict, pos_label='ad.')
cm_nb2 = confusion_matrix(y_test, y_predict, labels=['nonad.', 'ad.'])
a, b, _, _ = cm_nb2.ravel()
espefificidade_nb2 = a / (a + b)
cm_nb2 = pd.DataFrame(cm_nb2, index=['nonad.', 'ad.'], columns=['nonad.', 'ad.'])

cm_nb2
```
